# ðŸŽ¯ Disk Mesh Generation and Visualization Tutorial

In this tutorial, we'll learn how to use the [`disk_mesher_radial`](../src/surfmesh/disk.py#L9) and [`disk_mesher_square_centered`](../src/surfmesh/disk.py#L79) functions from the `surfmesh.disk` module to generate and visualize 2D disk meshes using Matplotlib.

We'll cover:
- Creating a radial disk mesh
- Creating a disk mesh with a square core
- Visualizing the meshes

## ðŸ“¦ 1. Import Dependencies

```python
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

from surfmesh import disk_mesher_radial, disk_mesher_square_centered
```

## ðŸŽ¡ 2. Generate a Radial Disk Mesh using [`disk_mesher_radial`](../src/surfmesh/disk.py#L9)

```python
# Parameters for the radial disk mesh
radius = 1.0
radial_resolution = 10
segment_resolution = 20

# Generate the radial disk mesh
radial_mesh = disk_mesher_radial(radius, radial_resolution, segment_resolution)

print(f"Generated radial disk mesh with {radial_mesh.shape[0]} quadrilateral faces.")
```

## ðŸ“Š 3. Visualize the Radial Disk Mesh

```python
plot_dir = Path("plots")
plot_dir.mkdir(exist_ok=True, parents=True)

fig, ax = plt.subplots(figsize=(6, 6))
patches = [Polygon(face, closed=True) for face in radial_mesh]
collection = PatchCollection(patches, facecolors="lightblue", edgecolors="k", alpha=0.7)
ax.add_collection(collection)

ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
ax.set_aspect("equal")
ax.set_title("Radial Disk Mesh")
fig.savefig(plot_dir / "radial_disk_mesh.jpg", dpi=150, bbox_inches="tight")
plt.show()
```

## ðŸŸ¦ 4. Generate a Disk Mesh with a Square Core using [`disk_mesher_square_centered`](../src/surfmesh/disk.py#L79)

```python
# Parameters for the square-centered disk mesh
# Explanation of parameters
# radius: The radius of the disk.
# square_resolution: The number of divisions along one side of the square core.
# radial_resolution: The number of radial divisions outside the square core.
# square_side_radius_ratio: The ratio of the square side length to the disk radius.

square_resolution = 5
radial_resolution = 10
square_side_radius_ratio = 0.5

# Generate the square-centered disk mesh
square_centered_mesh = disk_mesher_square_centered(radius, square_resolution, radial_resolution, square_side_radius_ratio)

print(f"Generated square-centered disk mesh with {square_centered_mesh.shape[0]} quadrilateral faces.")
```

## ðŸ“Š 5. Visualize the Square-Centered Disk Mesh

```python
fig, ax = plt.subplots(figsize=(6, 6))
patches = [Polygon(face, closed=True) for face in square_centered_mesh]
collection = PatchCollection(patches, facecolors="lightgreen", edgecolors="k", alpha=0.7)
ax.add_collection(collection)

ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
ax.set_aspect("equal")
ax.set_title("Square-Centered Disk Mesh")
fig.savefig(plot_dir / "square_centered_disk_mesh.jpg", dpi=150, bbox_inches="tight")
plt.show()
```

## âœ… Summary

- [`disk_mesher_radial`](../src/surfmesh/disk.py#L9) generates a radial disk mesh with curvilinear quadrilateral faces.
- [`disk_mesher_square_centered`](../src/surfmesh/disk.py#L79) generates a disk mesh with a square core and radial interpolation.
- Both meshes can be visualized using Matplotlib's `Polygon` and `PatchCollection`.
